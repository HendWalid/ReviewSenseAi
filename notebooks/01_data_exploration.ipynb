```python
# =========================================================
# ReviewSense AI - IMDb Data Exploration & Preprocessing
# =========================================================

# Load libraries
import pandas as pd

from datasets import load_dataset

from sklearn.feature_extraction.text import TfidfVectorizer

# import cleaning function
from preprocessing.clean_text import clean_text


# =========================
# Load IMDb Dataset
# =========================
dataset = load_dataset("imdb")

print(dataset)


# =========================
# Convert to DataFrames
# =========================
train_df = pd.DataFrame(dataset["train"])
test_df = pd.DataFrame(dataset["test"])


# =========================
# Display Sample
# =========================
print(train_df.head())

print("\nFirst Review:\n")
print(dataset["train"][0])


# =========================
# Apply Text Cleaning
# =========================
print("\nCleaning reviews...")

train_df["clean_review"] = train_df["text"].apply(clean_text)

test_df["clean_review"] = test_df["text"].apply(clean_text)


# =========================
# TF-IDF Feature Extraction
# =========================
print("\nExtracting TF-IDF features...")

vectorizer = TfidfVectorizer(max_features=5000)

X_train = vectorizer.fit_transform(train_df["clean_review"])

X_test = vectorizer.transform(test_df["clean_review"])


# =========================
# Labels
# =========================
y_train = train_df["label"]

y_test = test_df["label"]


# =========================
# Shapes
# =========================
print("\nTrain Features Shape:", X_train.shape)
print("Test Features Shape:", X_test.shape)

print("\nTrain Labels Shape:", y_train.shape)
print("Test Labels Shape:", y_test.shape)
```
